```python
#Questo algoritmo genera una catena lineare di agenti. Ne prende uno e ne genera un secondo con probabilità 
#inversamente proporzionale alla distanza da quello fissato. Infine mostra i valori della normalizzazione necessari
#per ottenere la distribuzione di probbabilità richiesta
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#definisco le lunghezze massima e minima per questa catena unidimensionale
Max=5001
Min=50
#creo una matrice che sarà il recipiente per i risultati della simulazione
info=np.zeros((Max-Min,2))
for N in range(Min,Max):
    a=np.array(range(1,N+1),dtype=np.float) #genero la catena
    invA=np.reciprocal(a) #genero la catena con gli indici reciproci
    hN=np.sum(invA) #calcola il valore ennesimo della serie armonica
    info[N-Min,0]=hN #metto questo calore tra i risultati
    prob=invA/hN#genero la distribuzione di probabilità
    num=1000
    elementA=np.random.randint(1,N+1, num)
    elementB=elementA+np.random.choice(a,size=num,p=prob)
    dist=elementB-elementA
    d=np.average(dist)
    info[N-Min,1]=d
plt.plot(info[:,1])
```
