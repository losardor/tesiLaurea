```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
Max=500
Min=50
info=np.zeros((Max-Min,2))
for N in range(Min,Max):
    a=np.array(range(1,N+1),dtype=np.float)
    invA=np.reciprocal(a)
    hN=np.sum(invA)
    info[N-Min,0]=hN
    prob=invA/hN
    num=100
    dist=np.zeros(num)
    for i in range(num):
        elementA=np.random.randint(1,N+1)
        elementB=elementA+np.random.choice(a,size=1,p=prob)
        dist[i]=elementB-elementA
    d=np.average(dist)
    info[N-Min,1]=d
plt.plot(info[:,1])
```
